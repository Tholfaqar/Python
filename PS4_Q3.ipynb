```python
def find_F(rho,rho_max,u_max,f):
    
    Rl = rho[:-1]
    Rr = rho[1:]
    f_rho = f(rho)
    fl = f_rho[:-1]
    fr = f_rho[1:]
    min_fvals = np.minimum(fl, fr)
    max_fvals = np.maximum(fl, fr)
    F = np.where(Rl < Rr, min_fvals, max_fvals)
    D = (Rr < 0.5 * rho_max) & (0.5 * rho_max < Rl)
    F[np.where(D)] = f(rho_max/2) 
    
    return F
```

```python
def calc_F2(F):
    F2 = F[1:] - F[:-1]
    return F2
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

rho_max = u_max = 1.0
rhoL = rho_max/2

dx = 4/400
dt = 0.8*dx/u_max
dt_dx = dt/dx
m = 400

x = dx * np.arange(1, m + 1, dtype=np.float64) - dx/2-2
f = lambda r: r * u_max * (1 - r / rho_max)
T = 2.0
T1 = 1.0
nsteps = int(round(T/dt))

rho = np.zeros((m))
rho = np.double(x < dx) *rhoL
rho0 = rho.copy()
rhof = np.zeros((10,m))
qdot = np.zeros((10))
tau = np.zeros((10))

qdot0 = 1.0
a = 1
# calculating the qdot
for k in range (0, 10):
    tau[k] = k*T/10
    qdot0 = 1.0
    rho = rho0.copy()
    a = 1
    while(qdot[k] - qdot0 <= 0.001):
        qdot0 = qdot[k]
        qdot[k] = 0.0
        for j in range (0,nsteps):
            F = find_F(rho,rho_max,u_max,f)
            # adding first traffic light at location x = 0.0 (i = 200 in x vector)
            if (j*dt < T1): 
                F[200] = 0.0
            # adding second traffic light at location x = 0.15 (i = 231 in x vector)
            if (tau[k] < T1) and (j*dt > tau[k] and j*dt < T1+tau[k]): 
                F[231] = 0.0
            if (tau[k] >= T1) and (j*dt < tau[k] and j*dt > tau[k]-T1):
                F[231] = 0.0
            rho[1:-1] = rho[1:-1] - dt_dx * (F[1:] - F[:-1])
            qdot[k] = qdot[k] + F[3*m/4]
        qdot[k] = qdot[k]/nsteps
        rhof[k,:] = rho[:]
        a = a +1
print(qdot)
plt.title('qdot vs tau - Gudonov scheme')
L1,=plt.plot(tau,qdot,label='qdot',lw=2)
Leg1=plt.legend(handles=[L1],loc=1)
plt.show()
plt.axes(xlim=(-2, 2), ylim=(0.0, 1.))
plt.title('rho final vs x - Godunovs scheme')
L1, =plt.plot(x,rhof[0],label='ru @ tau = 0.0',lw=1, color='k')
L2, =plt.plot(x,rhof[1],label='ru @ tau = 0.2',lw=1, color='g')
L3, =plt.plot(x,rhof[2],label='ru @ tau = 0.4',lw=1, color='r')
L4, =plt.plot(x,rhof[3],label='ru @ tau = 0.6',lw=1, color='c')
L5, =plt.plot(x,rhof[4],label='ru @ tau = 0.8',lw=1, color='m')
L6, =plt.plot(x,rhof[5],label='ru @ tau = 1.0',lw=1, color='y')
L7, =plt.plot(x,rhof[6],label='ru @ tau = 1.2',lw=1, color='k')
L8, =plt.plot(x,rhof[7],label='ru @ tau = 1.4',lw=1, color='g')
L9, =plt.plot(x,rhof[8],label='ru @ tau = 1.6',lw=1, color='r')
L10,=plt.plot(x,rhof[9],label='ru @ tau = 1.8',lw=1, color='c')

plt.show()
```
