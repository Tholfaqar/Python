$\frac{\partial{u}}{\partial{n}} = u_xn^x + u_yn^y = \frac{1}{J}[(y_\eta n^x - x_\eta n^y)u_\xi + (-y_\xi+ x_\xi n^y)u_\eta]$

descretizing $u_\xi $using central differences

$n^x = -y_\xi , n^y = x_\xi$

descretizing $u_\eta $using central differences

$u_{\eta_{0,j}} = \frac{u_{(0,j+1)}-u_{(0,j-1)}}{2h1}$

$u_{\xi_{(1,j)}} = \frac{-2}{h1} * (y_\eta^2 - x_\eta^2) $

$u_{\xi_{(2,j)}} = \frac{0.5}{h1} * (y_\eta^2 - x_\eta^2) $

```python
%matplotlib inline
import math
import numpy as np
import matplotlib.pyplot as plt
import scipy.sparse as sps
from scipy.sparse.linalg import spsolve
# FDPOISSON Solve Poisson's equation -(uxx + uyy) = f, bnd cnds u(x,y) = g(x,y)
# Problem parameters
n = 20
h1 = 1.0 / n
h = 1.0
b = 0.5
l = 3.0
d = 0.5 * (l - b)
a = 0.0
a = math.sqrt((0.25*(l - b)**2) - h**2)
a1 = 0.0
b1 = 0.0
c1 = 0.0
d1 = 0.0
e1 = 0.0
N = (n + 1)**2  # total number of grid points
# changing variable names to xi and eta
xi, eta = np.meshgrid(h1 * np.arange(n + 1), h1 * np.arange(n + 1))
x, y = np.meshgrid(0 * np.arange(n + 1), 0 * np.arange(n + 1))
f = lambda xi,eta: 1                            # sample right hand side
g = lambda xi,eta: 0*xi   # sample boundary conditions

# Mapping from grid indices to solution vector indices
umap = np.arange(N).reshape(xi.shape)

# Trick for boundary conditions: Start with Identity * u(x,y) = g(x,y)
A = sps.eye(N, N, format='lil')
K = g(xi.flatten(), eta.flatten())
```

```python
%matplotlib inline
# heat equation using Crank-Nicolson 
import math
import numpy as np
import scipy.sparse as sps
from scipy.sparse.linalg import spsolve
import matplotlib.pyplot as plt

# 1D heat equation, Crank-Nicolson scheme

#function: to solve heat equation using Crank-Nicolson method
def solveHeatEq_CN(m):
    
    h = 1.0 / (m+1)
    x = h * np.arange(m +1, dtype=np.float64)
    k = 0.1 * h
    T = 0.1
    nsteps = int(round(T/k))

    u = (np.sin(np.pi * x))**100 # Initial conditions
    # Dirichlet boundary conditions

    e = np.ones((m+1))
    A = 1 / h**2 * sps.spdiags(np.vstack([e, e, -2*e, e, e]), [-m, -1, 0, 1, m], m+1, m+1)
    I = sps.eye(m+1, m+1)
    LHSmatrix = I - k/2 * A
    RHSmatrix = I + k/2 * A

    for i in range(1,nsteps+1):
        u = spsolve(LHSmatrix, RHSmatrix.dot(u))
    
    return u
```

```python
# solving for multipe m values
m = 199
u = solveHeatEq_CN(m)
m = 99
u1 = solveHeatEq_CN(m)
m = 49
u2 = solveHeatEq_CN(m)
m = 24
u3 = solveHeatEq_CN(m)

# creating error matrices 
e1 = np.zeros(25, dtype=np.float64)
e2 = np.zeros(25, dtype=np.float64)
e3 = np.zeros(25, dtype=np.float64)
```
