euler_vortex

```python
import numpy as np
import scipy.sparse as sps

def euler_vortex(x, y, time, pars):
    gamma = 1.4
    rc, eps, M0, theta, x0, y0 = pars
    
    rinf = 1.0
    uinf = 1.0
    Einf = 1.0 / gamma / M0**2 / (gamma - 1) + 0.5
    pinf = (gamma - 1) * (Einf - 0.5)
    ubar = uinf * np.cos(theta)
    vbar = uinf * np.sin(theta)
    f = (1 - ((x - x0) - ubar * time)**2 \
           - ((y - y0) - vbar * time)**2) / rc**2

    u = uinf * (np.cos(theta) - eps * ((y - y0) - vbar * time)    \
                / (2 * np.pi * rc) * np.exp(f/2))
    v = uinf * (np.sin(theta) + eps * ((x - x0) - ubar * time)    \
                / (2 * np.pi * rc) * np.exp(f/2))
    r = rinf * (1 - eps**2 * (gamma - 1) * M0**2 / (8 * np.pi**2) \
                * np.exp(f))**(1.0 / (gamma - 1))
    p = pinf * (1 - eps**2 * (gamma - 1) * M0**2 / (8 * np.pi**2) \
                * np.exp(f))**(gamma / (gamma - 1))
    ru = r * u
    rv = r * v
    rE = p / (gamma - 1) + 0.5 * (ru**2 + rv**2) / r

    return r, ru, rv, rE
```

## a

```python
def euler_fluxes(r, ru, rv, rE):
    gamma = 7./5.
    p = (gamma-1)*r*((rE/r)-(((ru/r)**2+(rv/r)**2)/2))
    Frx = ru
    Fry = rv
    Frux = r*(ru/r)**2 + p
    Fruy = ru*(rv/r)
    Frvx = Fruy
    Frvy = (r*(rv/r)**2)+p
    FrEx = (ru/r)*(rE+p)
    FrEy = (rv/r)*(rE+p)
    return Frx, Fry, Frux, Fruy, Frvx, Frvy, FrEx, FrEy
```

## b

```python
import numpy as np
import scipy.sparse as sps
from scipy.sparse.linalg import spsolve

def compact_div(Fx,Fy,h):
    alpha = 0.25
    a = (2./3.)*(alpha+2)
    m = len(Fx)
    divFx=np.zeros((m,m))
    divFy=np.zeros((m,m))
    e = np.ones((m))
    z = np.zeros((m))
    LHS = sps.spdiags(np.vstack([alpha*e,alpha*e, e, alpha*e,alpha*e]), [-(m-1),-1, 0, 1,+(m-1)], m, m)
    RHS = (a/(2*h)) * sps.spdiags(np.vstack([e,-e,z,e,-e]), [-(m-1),-1, 0, 1,+(m-1)], m, m)
    divFx = (spsolve(LHS.tocsr(), RHS.dot(Fx.T))).T
    divFy = spsolve(LHS.tocsr(), RHS.dot(Fy))
    divF = divFx + divFy
    return divF
```

## c

```python
import numpy as np
import scipy.sparse as sps
from scipy.sparse.linalg import spsolve

def compact_filter(u, alpha):
    a = 5./8. + 3.*(alpha/4.)
    b = alpha +(1./2)
    c = (alpha/4)-(1./8)
    m = len(u)
    e = np.ones((m))
    RHS = sps.spdiags(np.vstack([(b/2)*e,(c/2)*e,(c/2)*e, (b/2)*e, a*e, (b/2)*e, (c/2)*e,(c/2)*e,(b/2)*e]), [-(m-1),-(m-2),-2,-1, 0, 1,2,+(m-2),+(m-1)], m, m)
    LHS = sps.spdiags(np.vstack([alpha*e, alpha*e, e, alpha*e, alpha*e]), [-(m-1),-1, 0, 1,+(m-1)], m, m)
    u1 = spsolve(LHS.tocsr(), RHS.dot(u))
    u2 = spsolve(LHS.tocsr(), RHS.dot(u1.T)).T
    return u2
```

## d

```python
def euler_rhs(r,ru,rv,rE,h):
    Frx, Fry, Frux, Fruy, Frvx,Frvy, FrEx, FrEy = euler_fluxes(r, ru, rv, rE)
    Fx = Frx
    Fy = Fry
    fr = -compact_div(Fx,Fy,h)
    Fx = Frux
    Fy = Fruy
    fru = -compact_div(Fx,Fy,h)
    Fx = Frvx
    Fy = Frvy
    frv = -compact_div(Fx,Fy,h)
    Fx = FrEx
    Fy = FrEy
    frE = -compact_div(Fx,Fy,h)
    
    return fr,fru,frv,frE
```

## e

```python
def euler_rk4step(r,ru,rv,rE,h,k,alpha):
    
    # rk4 scheme matrices 
    c1 = 0.0
    c2 = c3 = 0.5
    c4 = 1.
    b1 = b4 = 1./6.
    b2 = b3 = 1./3.
    a21 = a32 = a32 = 0.5
    a31 = a41 = a42 = 0.0
    a43 = 1.
    k11,k12,k13,k14 = euler_rhs(r,ru,rv,rE,h)
    k21,k22,k23,k24 = euler_rhs(r+(k/2)*k11,ru+(k/2)*k12,rv+(k/2)*k13,rE+(k/2)*k14,h)
    k31,k32,k33,k34 = euler_rhs(r+(k/2)*k21,ru+(k/2)*k22,rv+(k/2)*k23,rE+(k/2)*k24,h)
    k41,k42,k43,k44 = euler_rhs(r+k*k31,ru+k*k32,rv+k*k33,rE+k*k34,h)
    ur  = r  + k*(b1*k11+b2*k21+b3*k31+b4*k41)
    uru = ru + k*(b1*k12+b2*k22+b3*k32+b4*k42)
    urv = rv + k*(b1*k13+b2*k23+b3*k33+b4*k43)
    urE = rE + k*(b1*k14+b2*k24+b3*k34+b4*k44)
    
    r  = compact_filter(ur, alpha)
    ru = compact_filter(uru, alpha)
    rv = compact_filter(urv, alpha)
    rE = compact_filter(urE, alpha)
    
    return r,ru,rv,rE
```

## f

```python
%matplotlib inline
import numpy as np
import scipy.sparse as sps
import matplotlib.pyplot as plt

def calculate(N,alpha):
    T = 5.*np.sqrt(2)
    h = 10./N
    n = int(np.ceil(T/(0.3*h)))
    k = T/n
    x,y = np.meshgrid(h*np.arange(N+1),h*np.arange(N+1))
    pars = [0.5,1.0,0.5,np.pi/4,2.5,2.5]
    r,ru,rv,rE = euler_vortex(x,y,0.0,pars)
    for i in range(0,n):
        r,ru,rv,rE = euler_rk4step(r,ru,rv,rE,h,k,alpha)
    rfi,rufi,rvfi,rEfi = euler_vortex(x,y,T,pars)
    return r,ru,rv,rE, rfi,rufi,rvfi,rEfi
```

calculate errors

```python
# r is actual, rfi is r final at time = T etc..
def calc_err(r,ru,rv,rE,rfi,rufi,rvfi,rEfi):
    er=eru=erv=erE = 0.0
    Er=Eru=Erv=ErE = 0.0
    er=abs(r-rfi) #.sum(axis=1)
    eru=abs(ru-rufi) #.sum(axis=1)
    erv=abs(rv-rvfi) #.sum(axis=1)
    erE=abs(rE-rEfi) #.sum(axis=1)
    Er=np.max(er)
    Eru=np.max(eru)
    Erv=np.max(erv)
    ErE=np.max(erE)
    Er  = np.log(Er)
    Eru = np.log(Eru)
    Erv = np.log(Erv)
    ErE = np.log(ErE)
    return Er,Eru,Erv,ErE
```

alpha = .499 case

```python
%matplotlib inline
import numpy as np
import scipy.sparse as sps
from scipy.sparse.linalg import spsolve
import matplotlib.pyplot as plt

# alpha = 0.499 case
alpha = 0.499
N1 = 32
N2 = 64
N3 = 128
r,ru,rv,rE,rfi,rufi,rvfi,rEfi = calculate(N1,alpha)
er1,eru1,erv1,erE1 = calc_err(r,ru,rv,rE,rfi,rufi,rvfi,rEfi)

r,ru,rv,rE,rfi,rufi,rvfi,rEfi = calculate(N2,alpha)
er2,eru2,erv2,erE2 = calc_err(r,ru,rv,rE,rfi,rufi,rvfi,rEfi)

r,ru,rv,rE,rfi,rufi,rvfi,rEfi = calculate(N3,alpha)
er3,eru3,erv3,erE3 = calc_err(r,ru,rv,rE,rfi,rufi,rvfi,rEfi)

er  = ((er1,er2,er3))
eru = ((eru1,eru2,eru3))
erv = ((erv1,erv2,erv3))
erE = ((erE1,erE2,erE3))
dh = ((1./N1,1./N2,1./N3))
dh = np.log(dh)
print ('order of accuracy for r ',np.polyfit(dh,er,1))
print ('order of accuracy for ru',np.polyfit(dh,eru,1))
print ('order of accuracy for rv',np.polyfit(dh,erv,1))
print ('order of accuracy for rE',np.polyfit(dh,erE,1))

plt.title('Error vs. Grid Size @ alpha = 0.499')
L1,=plt.plot(dh,er,label='r err',lw=2)
L2,=plt.plot(dh,eru,label='ru err',lw=2)
L3,=plt.plot(dh,erv,label='rv err',lw=2)
L4,=plt.plot(dh,erE,label='rE err',lw=2)
Leg1=plt.legend(handles=[L1,L2,L3,L4],loc=2)
plt.show()
```

## g

```python
N = 256
T = 2
alpha = 0.48
gamma = 7./5
h = 1./N
n = int(np.ceil(T/(0.3*h)))
k = T/n
N = N+1
x,y = np.meshgrid(h*np.arange(N),h*np.arange(N))
rh  = lambda x,y: 2+x*0 if abs(y-0.5)<(0.15 + np.sin(np.pi*2*x)/200) else 1+x*0
r   = np.zeros((N,N))
for i in range(0,N):
    for j in range(0,N):
        r[i,j]=rh(x[i, j], y[i, j])

plt.figure()
plt.title('contour plot of initial r @ N = 256')
cset1 = plt.contourf(x, y, r, 10)
plt.colorbar()
plt.contour(x, y, r, cset1.levels, hold='on', colors='k')
plt.axis('equal')
plt.show()
I  = np.ones((N,N))
v  = np.zeros((N,N))
u  = r-I
ru = r*u
rv = r*v
p  = 3*I
E  = 1/(gamma-1)*(p/r)+(u**2+v**2)/2
rE = r*E

for i in range(0,n):
    r,ru,rv,rE = euler_rk4step(r,ru,rv,rE,h,k,alpha)

plt.figure()
plt.title('contour plot of final r @ N = 256')
cset1 = plt.contourf(x, y, r, 10)
plt.colorbar()
plt.contour(x, y, r, cset1.levels, hold='on', colors='k')
plt.axis('equal')
plt.show()
```
